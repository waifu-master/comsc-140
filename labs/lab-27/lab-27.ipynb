```python
import matplotlib.pyplot as plt 

file = open('example-1.csv', 'r')
db = file.readlines()
enrollments = []
years = []
for entry in db:
    data = entry.split(',')
    years.append(int(data[0]))
    enrollments.append(int(data[1]))

plt.title('DVC Enrollments 2012-2019')
plt.xlabel('Year')
plt.ylabel('# Enrolled')
plt.grid(True)
plt.plot(years, enrollments)
plt.show()
```
