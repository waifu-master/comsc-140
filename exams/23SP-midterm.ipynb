## Step 7: Add Division Functionality

Add a fourth possibility for a random math operation: division.

The user will have to enter their answer in whole number and fractional format, as show in this correct example:

    42 / 17 = 2 8/17

You'll need to figure out how to create the string "2 8/17" based on the two operands 42 and 17. You'll want to use string concatenation to build this string result. Remember that you can concatenate various elements together in a string like this example below. This example concatenates one value with a period and the other value, such as '10.42' or '162.42'. Don't forget to convert your numbers to strings first with str(), because you can only concatenate strings and string variables:

    result = str(val1) + '.' + str(val2)
    
How can you get the figures 2, 8, and 17 in the math example above?

    num1 = 42
    num2 = 17
    
    You can derive the "2" in "2 8/17" by "num1 // num2" (integer division).

    "17" is num2.

    Derive the "8" with num1 % num2 (modulus operator)

As a reminder, the integer division operator // will return just the integer portion of a division operation. These are true statements:

    42 // 17 = 2
    50 // 11 = 4

The modulus operator % will return just the remainder portion of a division operation. These are true statements:

    14 % 5 = 4  (5 goes evenly into 14 2 times with a remainder of 4)
    16 % 5 = 1  (5 goes evenly into 16 3 times with a remainder of 1)
    22 % 2 = 0  (2 goes evenly into 22 11 times with a remainder of 0)

The user should enter the string "2 8/17" including the space. Your answer string that you build should match that format.
