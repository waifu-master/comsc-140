```python
import os  # Needed for the remove and rename functions

def add_coffee():
    '''Adds a coffee record to coffee.txt'''
    # Create a variable to control the loop.
    another = 'y'

    # Open the coffee.txt file in append mode.
    coffee_file = open('coffee.txt', 'a')

    # Add records to the file.
    while another == 'y' or another == 'Y':
        # Get the coffee record data.
        print('Enter the following coffee data:')
        msg = 'Description: '
        descr = input(msg)
        print(msg, descr)
        msg = 'Quantity (in pounds): '
        qty = int(input(msg))
        print(msg, qty)

        # Append the data to the file.
        coffee_file.write(descr + '\n')
        coffee_file.write(str(qty) + '\n')

        # Determine whether the user wants to add
        # another record to the file.
        print('Do you want to add another record?')
        msg = 'Y = yes, anything else = no: '
        another = input(msg)
        print(msg, another)

    # Close the file.
    coffee_file.close()
    print('Data appended to coffee.txt.')

def delete_coffee():
    '''Delete coffee record from coffee.txt'''
    # Create a bool variable to use as a flag.
    found = False

    # Get the coffee to delete.
    msg = 'Which coffee do you want to delete? '
    search = input(msg)
    print(msg, search)
    
    # Open the original coffee.txt file.
    coffee_file = open('coffee.txt', 'r')

    # Open the temporary file.
    temp_file = open('temp.txt', 'w')

    # Read the first record's description field.
    descr = coffee_file.readline()

    # Read the rest of the file.
    while descr != '':
        # Read the quantity field.
        qty = float(coffee_file.readline())

        # Strip the \n from the description.
        descr = descr.rstrip('\n')

        # If this is not the record to delete, then
        # write it to the temporary file.
        if descr != search:
            # Write the record to the temp file.
            temp_file.write(descr + '\n')
            temp_file.write(str(qty) + '\n')
        else:
            # Set the found flag to True.
            found = True

        # Read the next description.
        descr = coffee_file.readline()

    # Close the coffee file and the temporary file.
    coffee_file.close()
    temp_file.close()

    # Delete the original coffee.txt file.
    os.remove('coffee.txt')

    # Rename the temporary file.
    os.rename('temp.txt', 'coffee.txt')

    # If the search value was not found in the file
    # display a message.
    if found:
        print('The file has been updated.')
    else:
        print('That item was not found in the file.')

def modify_coffee():
    '''Changes the quantity of some coffee record'''
    # Create a bool variable to use as a flag.
    found = False

    # Get the search value and the new quantity.
    msg = 'Enter a description to search for: '
    search = input(msg)
    print(msg, search)
    msg = 'Enter the new quantity: '
    new_qty = int(input(msg))
    print(msg, new_qty)
    
    # Open the original coffee.txt file.
    coffee_file = open('coffee.txt', 'r')

    # Open the temporary file.
    temp_file = open('temp.txt', 'w')

    # Read the first record's description field.
    descr = coffee_file.readline()

    # Read the rest of the file.
    while descr != '':
        # Read the quantity field.
        qty = float(coffee_file.readline())

        # Strip the \n from the description.
        descr = descr.rstrip('\n')

        # Write either this record to the temporary file,
        # or the new record if this is the one that is
        # to be modified.
        if descr == search:
            # Write the modified record to the temp file.
            temp_file.write(descr + '\n')
            temp_file.write(str(new_qty) + '\n')
            
            # Set the found flag to True.
            found = True
        else:
            # Write the original record to the temp file.
            temp_file.write(descr + '\n')
            temp_file.write(str(qty) + '\n')

        # Read the next description.
        descr = coffee_file.readline()

    # Close the coffee file and the temporary file.
    coffee_file.close()
    temp_file.close()

    # Delete the original coffee.txt file.
    os.remove('coffee.txt')

    # Rename the temporary file.
    os.rename('temp.txt', 'coffee.txt')

    # If the search value was not found in the file
    # display a message.
    if found:
        print('The file has been updated.')
    else:
        print('That item was not found in the file.')

def search_coffee():
    # Create a bool variable to use as a flag.
    found = False

    # Get the search value.
    search = input('Enter a description to search for: ')

    # Open the coffee.txt file.
    coffee_file = open('coffee.txt', 'r')

    # Read the first record's description field.
    descr = coffee_file.readline()

    # Read the rest of the file.
    while descr != '':
        # Read the quantity field.
        qty = float(coffee_file.readline())

        # Strip the \n from the description.
        descr = descr.rstrip('\n')

        # Determine whether this record matches
        # the search value.
        if descr == search:
            # Display the record.
            print('Description:', descr)
            print('Quantity:', qty)
            print()
            # Set the found flag to True.
            found = True

        # Read the next description.
        descr = coffee_file.readline()

    # Close the file.
    coffee_file.close()

    # If the search value was not found in the file
    # display a message.
    if not found:
        print('That item was not found in the file.')

def display_coffee():
    # Open the coffee.txt file.
    coffee_file = open('coffee.txt', 'r')

    # Read the first record's description field.
    descr = coffee_file.readline()

    # Read the rest of the file.
    while descr != '':
        # Read the quantity field.
        qty = float(coffee_file.readline())

        # Strip the \n from the description.
        descr = descr.rstrip('\n')

        # Display the record.
        print('Description:', descr)
        print('Quantity:', qty)

        # Read the next description.
        descr = coffee_file.readline()

    # Close the file.
    coffee_file.close()

def main():
    menu = '''***MAIN MENU***
[1] Display records
[2] Add record
[3] Search for record
[4] Modify record
[5] Delete record
[6] Quit'''
    divider = '-----------------------------------------------------------'
    prompt = 'Enter choice:'
    err_msg = 'ERROR: Input must be 1-6. Please enter another choice.'
    callback = [
        display_coffee,
        add_coffee,
        search_coffee,
        modify_coffee,
        delete_coffee
    ]
    choice = 0
    while choice != 6:
        print(menu)
        choice = int(input(prompt))
        while choice < 1 or choice > 6:
            choice = int(input(err_msg))
        print(prompt, choice)
        if choice == 6:
            break
        choice -= 1 # to fit the callback index
        callback[choice]()
        print(divider)

main()
```
