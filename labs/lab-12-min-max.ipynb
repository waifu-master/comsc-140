```python
def run_user_interface():
    '''
    Runs user interface
    '''
    main_menu = '''
Enter   1 to find min of two integers
        2 to find max of two integers
        3 to quit'''
    print(main_menu)

    # get mode
    msg = 'Your choice -->'
    mode = 0
    valid_modes = [1,2,3]
    while not(mode in valid_modes):
        mode = int(input(msg))
    print(msg, mode) # only print valid input\
    
    if mode == 3: # exit the UI immediately
        return False

    msg = 'Enter your two integers separated by a space: '
    num1, num2 = input(msg).split()
    num1 = int(num1)
    num2 = int(num2)
    print(f'You entered:\n\tValue 1: {num1}\n\tValue 2: {num2}')

    if mode == 1:
        chosen_val = findMin(num1, num2)
        if chosen_val == False:
            print('Results: The values are the same')
        else:
            print(f'Results: {chosen_val} is smaller')
    if mode == 2:
        chosen_val = findMax(num1, num2)
        if chosen_val == False:
            print('Results: The values are the same')
        else:
            print(f'Results: {chosen_val} is larger')
    
    return True

def findMax(num_1: int, num_2: int) -> int:
    '''
    get the max out of 2 numbers. will return False if same.
    '''
    if num_1 > num_2:
        return num_1
    elif num_2 > num_1:
        return num_2
    else:
        return False
    
def findMin(num_1, num_2) -> int:
    '''
    get the min out of 2 numbers. will return False if same.
    '''
    if num_1 < num_2:
        return num_1
    elif num_1 > num_2:
        return num_2
    else:
        return False

# main program
def main():
    is_running = True
    while (is_running):
        is_running = run_user_interface()
main()
```
