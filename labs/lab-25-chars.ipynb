```python
from typing import TypedDict

class Char_Stat(TypedDict):
    '''
    Stores how many characters
    '''
    count: int
    add_if: list[str]

def main():
    '''
    Entry point of the program
    '''
    msg = 'Enter a sentence'
    sentence = input(msg)
    print('Original sentence:')
    print(sentence)
    print()

    # calculate stats
    chars_count = len(sentence)
    # 1 space between each word, so 1 less than how many words there are
    spaces = len(sentence.split()) - 1 

    # made into a dict for extendability, in case more stats are needed
    stats = {
        'alphabetic':           Char_Stat(count=0, add_if=['lower', 'upper']),
        'alphanumeric':         Char_Stat(count=0, add_if=['lower', 'upper', 'digit']),
        'digits':               Char_Stat(count=0, add_if=['digit']),
        'lowercase letters':    Char_Stat(count=0, add_if=['lower']),
        'uppercase letters':    Char_Stat(count=0, add_if=['upper']),
        'other characters':     Char_Stat(count=0, add_if=['other'])
    }

    words = sentence.split()
    spaces = len(words) - 1
    for char in sentence:
        # categorize & count each stat
        if char == ' ':
            continue # skip spaces from counted as "other"
        cat = get_char_category(char)
        for key in stats:
            if cat in stats[key]['add_if']:
                stats[key]['count'] += 1
    
    # print the stats
    print(f'# of characters: {chars_count}')
    print(f'# of spaces: {spaces}')
    for key in stats:
        print(f'# of {key}: {stats[key]["count"]}')
    print()

    # print the modified sentences
    print(sentence.lower())
    print(sentence.upper())
    print(sentence.title())
    print("".join(reversed(sentence)))

def get_char_category(char: str) -> str:
    '''
    categorizes current character.
    can be lower, upper, digit, or other.
    '''
    if char.islower():
        return 'lower'
    elif char.isupper():
        return 'upper'
    elif char.isdigit():
        return 'digit'
    else:
        return 'other'

main()
```
