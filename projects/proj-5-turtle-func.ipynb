```python
import turtle
import re

def run_user_interface():
    '''
    Runs user interface
    '''
    main_menu = '''
    What would you like to do?
        1: draw a square
        2: draw a circle
        3: draw a line
        4: draw a triangle
        5: change the line width
        6: change the line color
        7: change the heading direction
        8: toggle pen up / down
        9: quit
    '''
    print(main_menu)

    # get mode
    mode = 0
    valid_modes = range(1, 10) # numbers 1 to 9, like the menu
    while not(mode in valid_modes):
        mode = int(input(main_menu))
    print(f'{main_menu}> {mode}') # only print valid input
    
    if mode == 9: # exit the UI immediately
        return False

    # numbers 1 to 5, which needs an extra size input
    size_input_modes = range(1, 6)
    if mode in size_input_modes:
        shapes = [
            'square',
            'circle',
            'line',
            'triangle',
            'line width'
        ]
        shape = shapes[mode - 1]
        size = get_size(shape)
        if shape != 'line width':
            draw_shape(shape, size)
        else:
            set_line_width(size)
    
    # the others need their own cases
    if mode == 6:
        color = input('Enter a color')
        set_line_color(color)
    
    if mode == 7:
        set_direction(int(input('Enter a direction')))
    
    if mode == 8:
        toggle_pen()
    
    return True

def get_size(shape: str):
    '''Get the object's size in px'''
    size = -1
    while size < 0:
        size = int(input(f"Enter the {shape}'s size in px"))
    return size

def set_line_color(color: str):
    '''Sets the line color with hex rgb codes'''
    turtle.pencolor(color)

def set_line_width(width: int):
    '''Sets turtle line width'''
    turtle.pensize(width)

def set_direction(direction: int):
    '''Sets turtle heading direction'''
    turtle.setheading(direction)

def toggle_pen():
    if turtle.isdown():
        turtle.penup()
    else:
        turtle.pendown()

def draw_shape(shape: str, size: int):
    '''Draws a shape with a specific size'''
    if shape == 'line':
        turtle.forward(size)
    if shape == 'circle':
        turtle.circle(size)
    if shape == 'triangle':
        for i in range(3):
            if (i != 0):
                turtle.left(120)
            turtle.forward(size)
    if shape == 'square':
        for i in range(4):
            if (i != 0):
                turtle.left(90)
            turtle.forward(size)

# main program
def main():
    turtle.setup(600, 600)
    is_running = True
    while (is_running):
        is_running = run_user_interface()
    turtle.done()
main()
```
