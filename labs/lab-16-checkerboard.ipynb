```python
import turtle
NUM_SQUARES = 5
MIN_WIDTH = 30
MAX_WIDTH = 50

def square(width, color):
    '''Draws a square with specified width & color'''
    turtle.fillcolor(color)
    turtle.begin_fill()
    for count in range(4):
        turtle.forward(width)
        turtle.left(90)
    turtle.forward(width)
    turtle.end_fill()

def main():
    '''Entry point of the program'''
    colors = [
        'black',
        'maroon',
        'midnightblue',
        'darkslategray'
    ]

    colors_disp = [
        'Black',
        'Maroon',
        'Midnight Blue',
        'Dark Slate Gray'
    ]

    # enter the color
    prompt = 'Choose a color'
    for i in range(len(colors_disp)):
        prompt += f'\n[{i + 1}] {colors_disp[i]}'

    color_choice = int(turtle.numinput(
        'Input', 
        prompt,
        minval = 1,
        maxval = len(colors)
    )) - 1 # option starts at 1, array access starts at 0, so must be offset
    color = colors[color_choice]

    size = int(turtle.numinput(
        'Input', 
        'Enter the square size [30-50]',
        minval = MIN_WIDTH,
        maxval = MAX_WIDTH
    ))

    # ideally this should be 0 but it wouldn't fit in my screen
    # so i had to put in buffers & adjust their values by tests
    scr_buffer = 30
    starting_pos_buffer = 5
    ideal_screen_size = size * NUM_SQUARES
    screen_size = ideal_screen_size + scr_buffer
    ideal_starting_pos = -(ideal_screen_size / 2) # the edge of the ideal screen
    starting_pos_x = ideal_starting_pos - starting_pos_buffer # need a left nudge
    starting_pos_y = ideal_starting_pos + starting_pos_buffer # need a up nudge
    
    # set up the checker board
    turtle.setup(screen_size, screen_size)
    turtle.penup()
    turtle.goto(starting_pos_x, starting_pos_y)
    turtle.speed(1000)

    # draw the checker board
    square_num = 0 # color if even, white if odd to keep alternating pattern
    for row in range(NUM_SQUARES):
        # reset the position of the pen to left edge of current row
        turtle.goto(starting_pos_x, starting_pos_y + row * size)

        # draw squares in this row left to right
        for col in range(NUM_SQUARES): 
            if (square_num % 2 == 0): # fill an even square
                turtle.pendown()
                square(size, color)
                turtle.penup()
            else:
                turtle.pendown()
                square(size, 'white')
                turtle.penup()
            square_num += 1

    turtle.done()

main()
```
