```python
def is_prime(val):
    if val < 2:
        return False
    for i in range(2, int(val/2)+1):
        if val % i == 0:
            return False
    return True

def main():
    filename = 'numbers.txt'
    file = open(filename, 'r')

    counter = 0
    global_sum = 0
    global_average = 0

    counter_even = 0
    counter_odd = 0
    max_num = -1
    min_num = -1

    counter_prime = 0
    sum_prime = 0
    average_prime = 0
    # holder values for undefined primes
    max_prime = -1
    min_prime = -1

    for number in file:
        number = int(number)
        counter += 1
        global_sum += number
        global_average = global_sum / counter

        if number % 2 == 0:
            counter_even += 1
        else:
            counter_odd += 1

        if number > max_num or max_num == -1:
            max_num = number
        if number < min_num or min_num == -1:
            min_num = number

        if is_prime(number):
            counter_prime += 1
            sum_prime += number
            average_prime = sum_prime / counter_prime
            if number > max_prime or max_prime == -1:
                max_prime = number
            if number < min_prime or min_prime == -1:
                min_prime = number
        
    pct_even = round(counter_even * 100 / counter, 2)
    pct_odd = round(counter_odd * 100 / counter, 2)
    pct_prime = round(counter_prime * 100 / counter, 2)
    global_average = round(global_average, 2)
    average_prime = round(average_prime, 2)
    print(f''' 
Count of numbers:         {counter:>15,}
Sum of numbers:           {global_sum:>15,}
Average of numbers:       {global_average:>15,}
Count of even numbers:    {counter_even:>15,} ({pct_even}%)
Count of odd numbers:     {counter_odd:>15,} ({pct_odd}%)
Highest number:           {max_num:>15,}
Lowest number:            {min_num:>15,}
Count of prime numbers:   {counter_prime:>15,} ({pct_prime}%)
Sum of prime numbers:     {sum_prime:>15,}
Average of prime numbers: {average_prime:>15,}
Highest prime number:     {max_prime:>15,}
Lowest prime number:      {min_prime:>15,}
    ''')

main()
```
