```python
import itertools # standard iteration tool to iterate every combination

def main():
    '''Entry point'''
    characters = ['X', 'Q', 'Z', '&', '$', 'K']
    letter_count = 3
    max_col = 10

    # this gets every word, with repeated characters included
    words = [word for word in itertools.product(characters, repeat = letter_count)]

    # print the columns
    row = ''
    counter = 0
    for word_tuple in words:
        word = ''
        for char in word_tuple:
            word += char
        row += word + '  '
        counter += 1
        if counter >= max_col:
            print(row)
            row = ''
            counter = 0

main()
```
